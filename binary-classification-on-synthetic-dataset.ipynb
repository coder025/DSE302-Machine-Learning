# Binary classification on synthetic dataset


```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import cross_validate
from sklearn.model_selection import KFold
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.model_selection import GridSearchCV
```

```python
import warnings
warnings.filterwarnings("ignore")
```

Reading CSV Data

```python
X = pd.read_csv('/content/training_data.csv',header=None)
X_test = pd.read_csv('/content/test_data.csv',header=None)
y = pd.read_csv('/content/training_data_class_labels.csv',header=None)
```

```python
X.plot.scatter(x=0,y=1)
```

```python
X_test.plot.scatter(x=0,y=1)
```

```python
X_train,X_val,y_train,y_val= train_test_split(X,y,test_size=0.111)
```

```python
knn = KNeighborsClassifier()
knn.fit(X_train,y_train)
y_pred = knn.predict(X_test)
print(classification_report(y_val,y_pred))
```

Logistic Regression without HyperParameter Tuning

```python
lr = LogisticRegression()
lr.fit(X_train,y_train)
lr.score(X_val,y_val)
```

Logistic Regression after HyperParameter Tuning

```python
# Find best hyperparameters (accuracy)
log_clf = LogisticRegression()
param_grid = {
            'penalty' : ['l2','l1','none','elasticnet'],  
            'C' : [0.001, 0.01, 0.1, 1, 10, 100,1000],
            'max_iter':[100,200,300],
            'verbose':[0,5,10,12,15]
    
            }

CV_log_clf = GridSearchCV(estimator = log_clf, param_grid = param_grid , scoring = 'accuracy', verbose = 1, n_jobs = -1,)
CV_log_clf.fit(X_train, y_train)

best_parameters = CV_log_clf.best_params_
print('The best parameters for using this model is', best_parameters)
```

```python
CV_log_clf = LogisticRegression(C = best_parameters['C'], 
                                penalty = best_parameters['penalty'], 
                                random_state = 40)

CV_log_clf.fit(X_train, y_train)
y_pred = CV_log_clf.predict(X_test)
CV_log_clf.score(X_val,y_val)

```

```python
print(classification_report(y_val, y_pred))
```

KNN


```python
knn = KNeighborsClassifier()
knn.fit(X_train,y_train)
y_pred = knn.predict(X_test)
knn.score(X_val,y_val)
```

```python
print(classification_report(y_val, y_pred))
```

Naive Bayes Classifier


```python
nb= GaussianNB()
nb.fit(X_train,y_train.values.ravel())
y_pred =nb.predict(X_val)
accuracy_score(y_val,y_pred)
print(classification_report(y_val, y_pred))
```

```python
# Find best hyperparameters (accuracy)
nb_clf = GaussianNB()
param_grid = {'var_smoothing':[1e-9,1e-6,1e-3,1e-2],
          }

CV_nb_clf = GridSearchCV(estimator = nb_clf, param_grid = param_grid , scoring = 'accuracy', verbose = 1)
CV_nb_clf.fit(X_train, y_train)

best_parameters = CV_nb_clf.best_params_
print('The best parameters for using this model is', best_parameters)
```

```python
CV_nb_clf = GaussianNB(var_smoothing = best_parameters['var_smoothing'], 
                                 )

CV_nb_clf.fit(X_train, y_train)
CV_nb_clf.score(X_val,y_val)
y_pred = CV_nb_clf.predict(X_test)
```

```python
print(classification_report(y_val, y_pred))
```
